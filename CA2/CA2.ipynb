#### For embarked column:
We replace the missing values in the 'embarked' column with the most common value because there are only two missing values. 

```python
modeEmbarked = dataFrameTaskOne["embarked"].mode()[0]

dataFrameTaskOne["embarked"] = dataFrameTaskOne["embarked"].replace(-1, modeEmbarked)
```

#### For age column:
First, we determined the age range for each age group (man, woman, and child). Then, we filled missing values with a random age selected from within the appropriate age ranges.

```python
#We found these ranges according to the 'who' and 'age' columns:
ageRanges = {1: (16, 80), 2: (16, 63), 3: (1, 15)}

def fillAge(row):
    if row["age"] == -1:
        if row["who"] in ageRanges:
            minAge, maxAge = ageRanges[row["who"]]
            return np.random.randint(minAge, maxAge)
    else:
        return row["age"]
    

dataFrameTaskOne["age"] = dataFrameTaskOne.apply(fillAge, axis=1)
```

#### For deck column:


At first, we wanted to find a relationship between the 'pclass' and 'deck' columns because the 'deck' column had the highest correlation with the 'pclass' column. However, that approach did not go well. Therefore, we decided to fill the missing values in the 'deck' column with a random sample generated according to the non-missing values in the same column.

```python
def printDataFrequencyOfDeck(pclass):
    classData = dataFrameTaskOne[dataFrameTaskOne["pclass"] == pclass]
    return classData["deck"].value_counts()

print("First Class:", printDataFrequencyOfDeck(1))
print("Second Class:", printDataFrequencyOfDeck(2))
print("Third Class:", printDataFrequencyOfDeck(3))
```

```python
nonMissingDeck = dataFrameTaskOne["deck"][dataFrameTaskOne["deck"] != -1]

randomSample = nonMissingDeck.sample(n=len(dataFrameTaskOne[dataFrameTaskOne["deck"] == -1]), replace=True)

dataFrameTaskOne.loc[dataFrameTaskOne["deck"] == -1, "deck"] = randomSample.values
```
